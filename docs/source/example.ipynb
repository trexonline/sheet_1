# Example code for simulator

```python
from qiskit import QuantumCircuit, transpile
import matplotlib.pyplot as plt
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import QFTGate
from sheet_1.quantum_simulator import simulate



qc = QuantumCircuit(4)
'''Quantum circuit with 4 qubits'''

qc.h(0) 
qc.h(1)
qc.h(2)
qc.h(3)
'''Hardermard gates, |0>^4 -> 1/4*(|0> + |1>)^4 state'''

qc.append(QFTGate(4), [0, 1, 2, 3])
transpiled_qc = transpile(qc, basis_gates=["cx", "u"])
'''transpiles quantum cirquit to one using only "u" and "cx" gates'''


result = simulate(transpiled_qc)


print(result)

'''gives the statevector'''

transpiled_qc.draw("mpl")

'''draws a graphic of the gates'''


plt.show()
```
